```python
# importing libraries


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




# DATA SETS OF PATIENTS


data={
    "patient":["Ali","Sara","Ahmad","Ayesha","Bilal","Usman"],
    "Age":[45,50,38,60,42,35],
    "pain_level":[7,5,8,6,4,9],
    "Sessions":[10,8,12,9,7,14],
    "Recovery_status":[
        "Need Attention",
        "improving",
        "Need Attention",
        "Need Attention",
        "Improving",
        "Need Attention"
    ]

}


# DATA FRAME


df=pd.DataFrame(data)
print(df)

# SEABORN STYLE
sns.set_style("darkgrid")

# BARPLOT
sns.barplot(x="patient",y="pain_level",data=df)

plt.title("patient pain levels")
plt.xlabel("patients")
plt.ylabel("pain_level")
plt.show()


# LINE PLOT

sns.lineplot(x="patient",y="Sessions",data=df,marker="o")
plt.title("Theory sessions")
plt.xlabel("patients")
plt.ylabel("sessions")
plt.show()



# COUNT PLOT (RECOVERY STATUS)

sns.countplot(x="Recovery_status",data=df)
plt.title("Recovery status count")
plt.show()


# HEATMAP

correlation=df[["pain_level","Sessions","Age"]].corr()
sns.heatmap(correlation,annot=True)
plt.title("Correlation Heatmap")
plt.show()


# CSV FILE SAVE

df.to_csv("physio_visualization_project.csv",index=False)
```
